```python
import cv2
from ultralytics import YOLO
from collections import defaultdict

#Load YOLO Model
model=YOLO('yolo11l.pt')
class_list = model.names 
# open the video
cap=cv2.VideoCapture('test_videos/4.mp4')
while cap.isOpened():
    ret, frame=cap.read()
    if not ret:
        break
    # run yolo tracking on the frame
    results=model.track(frame,persist=True, classes = [1,2,3,5,6,7])

    if results[0].boxes.data is not None:
        boxes=results[0].boxes.xyxy.cpu()
        track_ids=results[0].boxes.id.int().cpu().tolist()
        class_indices=results[0].boxes.cls.int().tolist()
        confidences=ces=results[0].boxes.conf.cpu()

        for box , track_id, class_id, conf in zip(boxes,track_ids,class_indices,confidences):
            x1,y1,x2,y2 =map(int,box)
            class_name=class_list[class_id]
    
    
            cv2.putText(frame,f"ID: {track_id} {class_name}",(x1,y1-10),cv2.FONT_HERSHEY_SIMPLEX,0.6,(0,255,255),2)          
            cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2)

    #show the frame
    cv2.imshow('Yolo objetc tracking et counting',frame)

    if cv2.waitKey(0) & 0xff == ord('q'):
        break
    
# Release ressource
cap.release()
cv2.destroyAllWindows()




```
